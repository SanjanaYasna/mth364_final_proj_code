```python
#iterate through df
with open("nf1_kma_final.csv", "w") as f:   
    f.write("k_ma,A,B,trace,determinant,stable,string_of_classification\n")
    for index, row in df.iterrows():
        #get tracea dn determinant
        trace = row['trace']
        det = row['determinant']
        if det < 0:
            classification = "unstable saddle node"
            stable = False
        elif trace > 0 and trace**2 < 4*det:
            classification = "unstable spiral node source"
            stable = False
        elif trace > 0 and trace**2  > 4*det:
            classification = "unstable node source"
            stable = False   
        elif trace < 0 and trace**2 > 4*det:
            classification = "stable spiral node sink"
            stable = True
        else:
            classification = "stable node sink"
            stable = True
        f.write(f"{row['k_ma']},{row['A']},{row['B']},{trace},{det},{stable},{classification}\n")
        f.flush()
# f.close()
# for index, row in df.iterrows():
#     #get tracea dn determinant
#     trace = row['trace']
#     det = row['determinant']
#     if det < 0:
#         classification = "unstable saddle node"
#         stable = False
#     elif trace > 0 and trace**2 < 4*det:
#         classification = "unstable spiral node source"
#         stable = False
#     elif trace > 0 and trace**2  > 4*det:
#         classification = "unstable node source"
#         stable = False   
#     elif trace < 0 and trace**2 > 4*det:
#         classification = "stable spiral node sink"
#         stable = True
#     else:
#         classification = "stable node sink"
#         stable = True
#     if stable != row['stable']:
#         print(f"Row {index} has a mismatch in stability classification.")
#         print(f"Trace: {trace}, Determinant: {det}, Stable: {stable}, Classification: {classification}")
#     if classification != row['string_of_classification']:
#         print(f"Row {index} has a mismatch in classification.")
#         print(f"Trace: {trace}, Determinant: {det}, Stable: {stable}, Classification: {classification}")
    
```
